```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv('output.csv')

# Rename the columns for consistency
df.columns = df.columns.str.strip()

# Count the occurrences of each process name
process_count = df['Process Name'].value_counts().head(10)

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 16))

# Plot the bar chart
sns.barplot(x=process_count.values, y=process_count.index, palette='viridis', ax=axes[0])
axes[0].set_xlabel('Number of Occurrences')
axes[0].set_ylabel('Process Names')
axes[0].set_title('Top 10 Processes with Most Occurrences')

# Plot the pie chart
axes[1].pie(process_count, labels=process_count.index, autopct='%1.1f%%', colors=sns.color_palette('viridis'), startangle=90)
axes[1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
axes[1].set_title('Top 10 Processes with Most Occurrences')

# Adjust layout
plt.tight_layout()

plt.show()
```
