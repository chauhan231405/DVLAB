##8 Visual Variables
1. Position
2. Mark/ Shape
3. Size
4. Brightness/ Saturation
5. Color
6. Orientation/ Direction
7. Texture
8. Motion

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Read the dataset
df = pd.read_csv('HousingData.csv')

# Replace NaN values with the mean of the column
df = df.fillna(df.mean())
```

```python
# Visual Variable 1: Position
# Scatter plot between two variables (e.g., RM and MEDV)
plt.scatter(df['RM'], df['MEDV'])
plt.xlabel('Average Number of Rooms per Dwelling (RM)')
plt.ylabel('Median Value of Owner-Occupied Homes (MEDV)')
plt.title('Scatter Plot: Position')
plt.show()
```

```python
# Visual Variable 2: Mark or Shape
# Scatter plot between two variables (e.g., RM and MEDV) using different marks for 'RAD' categories
plt.figure(figsize=(10, 6))
markers = ['o', 's', '^', 'D', 'v', '<', '>', 'p']  # Different markers for each 'RAD' category
for rad, marker in zip(df['RAD'].unique(), markers):
    subset = df[df['RAD'] == rad]
    plt.scatter(subset['RM'], subset['MEDV'], marker=marker, label=f'RAD {rad}', alpha=0.7)
plt.xlabel('Average Number of Rooms per Dwelling (RM)')
plt.ylabel('Median Value of Owner-Occupied Homes (MEDV)')
plt.title('Scatter Plot: Mark or Shape')
plt.legend()
plt.show()
```

```python
# Visual Variable 3: Size
# Bubble plot to represent a third dimension (e.g., RM, MEDV, and LSTAT)
plt.scatter(df['RM'], df['MEDV'], s=df['LSTAT']*10, alpha=0.5)
plt.xlabel('Average Number of Rooms per Dwelling (RM)')
plt.ylabel('Median Value of Owner-Occupied Homes (MEDV)')
plt.title('Bubble Plot: Size')
plt.show()
```

```python
# Visual Variable 4: Brightness/Saturation
# Scatter plot with brightness variation (e.g., RM and DIS)
plt.scatter(df['RM'], df['DIS'], c=df['DIS'], cmap='gray', alpha=0.7)
plt.xlabel('Average Number of Rooms per Dwelling (RM)')
plt.ylabel('Weighted Distances to Employment Centers (DIS)')
plt.title('Scatter Plot: Brightness/Saturation')
plt.colorbar(label='DIS')
plt.show()
```

```python
# Visual Variable 5: Color
# Scatter plot with color variation (e.g., RM and AGE)
plt.scatter(df['RM'], df['AGE'], c=df['AGE'], cmap='viridis', alpha=0.7)
plt.xlabel('Average Number of Rooms per Dwelling (RM)')
plt.ylabel('Proportion of Owner-Occupied Units Built Prior to 1940 (AGE)')
plt.title('Scatter Plot: Color')
plt.colorbar(label='AGE')
plt.show()
```

```python
# Visual Variable 6: Orientation or Direction
# Scatter plot with arrows indicating orientation based on TAX values
plt.figure(figsize=(10, 6))
# Define arrow orientations based on TAX values
arrow_orientations = df['TAX'] * (2 * np.pi / df['TAX'].max())
# Scatter plot with arrows indicating orientation
plt.scatter(df['RM'], df['MEDV'], alpha=0.7)
for x, y, orientation in zip(df['RM'], df['MEDV'], arrow_orientations):
    plt.arrow(x, y, np.cos(orientation), np.sin(orientation), width=0.1, color='blue', alpha=0.7)
plt.xlabel('Average Number of Rooms per Dwelling (RM)')
plt.ylabel('Median Value of Owner-Occupied Homes (MEDV)')
plt.title('Scatter Plot: Arrow Orientation based on TAX')
plt.show()
```

```python
# Visual Variable 7: Texture
# Scatter plot to show different textures for each 'RAD' category with larger block sizes
plt.figure(figsize=(18, 12))
# Using valid marker styles recognized by Matplotlib
textures = ['/', '\\', '|', '-', '+', 'x', 'o', '*']
for i, rad in enumerate(df['RAD'].unique()):
    subset = df[df['RAD'] == rad]
    plt.scatter(subset['RM'], subset['MEDV'], marker='s', s=500, label=f'Category RAD {rad}', alpha=0.7, hatch=textures[i % len(textures)])
plt.xlabel('Average Number of Rooms per Dwelling (RM)')
plt.ylabel('Median Value of Owner-Occupied Homes (MEDV)')
plt.title('Scatter Plot: Texture')
plt.legend()
plt.show()
```

```python
# Visual Variable 8: Motion
# Create an interactive 3D scatter plot with plotly
fig = px.scatter_3d(df, x='RM', y='MEDV', z='LSTAT', opacity=0.7,
                    title='Interactive 3D Scatter Plot',
                    labels={'RM': 'Average Number of Rooms',
                            'MEDV': 'Median Home Value',
                            'LSTAT': 'Lower Status of the Population'})
# Adjust the size of the canvas
fig.update_layout(width=1000, height=800)
# Show the plot
fig.show()
```
